```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
oliv = datasets.fetch_olivetti_faces()
```

```python
oliv.keys()
```

```python
oliv['data'].shape
```

```python
oliv['images'].shape
```

```python
fig = plt.figure(figsize = (8,8))
for i in range(64):
    ax = fig.add_subplot(8,8,i+1) # row, column, index
    ax.imshow(oliv['images'][i], cmap = plt.cm.bone)
plt.show()
```

```python
x = oliv.data
y = oliv.target
```

```python
pca = PCA()
pca.fit(x)
```

```python
pca.components_.shape
```

```python
k = 0 
current_variance = 0
total = sum(pca.explained_variance_)
while current_variance/total < 0.95:
    current_variance += pca.explained_variance_[k]
    k += 1
k
```

```python
pca = PCA(n_components = k, whiten = True)
transfromed_data = pca.fit_transform(x)
transfromed_data.shape
```

```python
# restore the images from transformed data which lower dimensional data
x_approx = pca.inverse_transform(transfromed_data)
x_approx.shape
```

```python
x_approx_images = x_approx.reshape((400,64,64))
```

```python
fig = plt.figure(figsize = (8,8))
for i in range(64):
    ax = fig.add_subplot(8,8,i+1) # row, column, index
    ax.imshow(x_approx_images[i], cmap = plt.cm.bone)
plt.show()
```

```python
eigen_vec = pca.components_
eigen_vec.shape
```

```python
eigen_faces = eigen_vec.reshape((123,64,64))
```

```python
fig = plt.figure(figsize = (8,8))
for i in range(64):
    ax = fig.add_subplot(8,8,i+1) # row, column, index
    ax.imshow(eigen_faces[i], cmap = plt.cm.bone)
plt.show()
```
