```python
import numpy as np
```

Basics of Array

```python
# 1D array
arr = np.array([1,2,3,6])
print(arr)
```

```python
arr.shape
```

```python
# 2D array as float data type
arr2 = np.array([(1,2,3), (5,6,7)], dtype = float)
print(arr2)
```

```python
arr2.shape
```

```python
# array of zeros
zero = np.zeros((3,3))
print(zero)
```

```python
# array of particular value
a = np.full((3,4), 5)
print(a)
```

```python
# array of evenly spaced values
b = np.linspace(10,30,5)
print(b)
```

```python
# array of evenly spaced values specifying a step value
c = np.arange(10,30,5)
print(c)
```
