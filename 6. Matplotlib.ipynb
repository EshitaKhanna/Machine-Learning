```python
import matplotlib.pyplot as plt
```

```python
# scatter plot
x = [1,2,3]
y = [2,4,6]
plt.scatter(x,y)
plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]

plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

```python
# plot x^3
import numpy as np
#x = np.array([1,2,3,4])
x = np.arange(0,5,0.1)
y = x**3
plt.plot(x,y)
plt.show()
```

```python
a = [3,4,5] # treats this as y 
plt.plot(a, "b--")
plt.show()
```

```python
x = np.arange(0,5,0.1)
y = x**3
y2 = x**2

#plt.plot(x,y, color = "blue", marker = "+")
plt.plot(x, y, color = "black", linewidth = 5, label = "x^3")
plt.plot(x, y2, color = "red", linewidth = 5, label = "x^2")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Powers of x")

plt.legend()
plt.grid()

plt.text(2,80, "Text", fontsize = 12)
plt.show()
```

```python
# Pie Chart
labels = ["A", "B", "C", "D"]
sizes = [3,4,6,2]
colours = ["yellow", "purple", "red", "green"]
explode = [0.1, 0, 0, 0]

plt.pie(sizes, colors = colours, labels = labels, autopct = "%.2f%%", explode = explode, counterclock = False, startangle = 100)
plt.axis("equal")
plt.show()
```
