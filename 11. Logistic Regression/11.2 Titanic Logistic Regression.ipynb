```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("training_titanic_x_y_train.csv")
```

```python

df.head()
```

```python
df.drop(["Name","Ticket"],axis=1,inplace=True)
```

```python
def gender(g):
    if g == "male":
        return 1
    else:
        return 0
df["Gender"] = df["Sex"].apply(gender)
df.head()
```

```python
df.drop(['Sex'], axis = 1, inplace =True)
df.head()
```

```python
df.isnull().sum()
```

```python
def cabin(x):
    if x is np.nan:
        return 0
    else:
        return 1
df["Cabin"] = df.Cabin.apply(cabin)
df.head()
```

```python
df.isnull().sum()
```

```python
# Fill mean Survived age in the column where the person has survived and 
# mean not Survived age in the column where person has not survived

mean_age_sur = df[df.Survived == 1].Age.mean()
mean_age_not_sur = df[df.Survived == 0].Age.mean()

df.Age.fillna(mean_age_sur, inplace = True)
df.Age.fillna(mean_age_not_sur, inplace = True)
df.head()
```

```python
df.isnull().sum()
```

```python
def cherbourg(s):
    if s=="C":
        return 1
    else:
        return 0
    
def queenstown(s):
    if s=="Q":
        return 1
    else:
        return 0

def southamptom(s):
    if s=="S":
        return 1
    else:
        return 0
        
df["Cherbourg"] = df.Embarked.apply(cherbourg)
df["Queenstown"] = df.Embarked.apply(queenstown)
df["Southamptom"] = df.Embarked.apply(southamptom)
df.drop("Embarked",axis=1, inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df = df[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare','Cabin','Gender','Cherbourg','Queenstown','Southamptom','Survived']]
df.head()
```

```python
data = df.values
data
```

```python
data.shape
```

```python
X = data[:, 0:10]
Y = data[:, 10]
X.shape,Y.shape
```

```python
clf = LogisticRegression(max_iter=1000)
```

```python
clf.fit(X, Y)
```

```python
clf.score(X, Y)
```

```python
clf.predict_proba(X)
```

```python
clf.predict(X) - Y
```
