```python
import numpy as np
import pandas as pd
from tensorflow.keras.datasets import cifar10
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
(x_train,y_train), (x_test, y_test) = cifar10.load_data()
```

```python
x_train.shape, x_test.shape
```

```python
# classes
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
y_train = y_train.flatten()
y_test = y_test.flatten()
```

```python
fig = plt.figure(figsize = (12,10))
for i in range(36):
    ax = fig.add_subplot(6,6,i+1) # row, column, index
    ax.imshow(x_train[i], cmap = plt.cm.bone)
    ax.set_title(class_names[y_train[i]])
    ax.set_xticks([])
    ax.set_yticks([])
plt.show()
```

```python
# flattening the data
x_train_flat = x_train.reshape(x_train.shape[0],-1)
x_test_flat = x_test.reshape(x_test.shape[0],-1)

# normalizing the data
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train_flat)
x_test_scaled = scaler.transform(x_test_flat)
```

```python
pca = PCA()
pca.fit(x_train_scaled)
pca.components_.shape
```

```python
k = 0 
current_variance = 0
total = sum(pca.explained_variance_)
while current_variance/total < 0.95:
    current_variance += pca.explained_variance_[k]
    k += 1
k
```

```python
pca = PCA(n_components = k, whiten = True)
x_train_pca = pca.fit_transform(x_train_scaled)
x_train_pca.shape
```

```python
x_test_pca = pca.transform(x_test_scaled)
```

#### Using Random Forest

```python
rf = RandomForestClassifier(n_estimators=500, max_depth=50, min_samples_split=2, random_state=42)
rf.fit(x_train_pca, y_train)
predictions = rf.predict(x_test_pca)
```

```python
predictions
```

```python
prediction_names = []
for i in predictions:
    prediction_names.append(class_names[i])
prediction_names
```

```python
print(accuracy_score(y_test, predictions))
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```

```python
np.savetxt('cifar10_predictions.csv', prediction_names, fmt='%s', delimiter=',')
```

#### Using SVM

```python
svm = SVC(kernel='linear', random_state=42)
svm.fit(x_train_pca, y_train)
predictions_svm = svm.predict(x_test_pca)

prediction_names = []
for i in predictions_svm:
    prediction_names.append(class_names[i])

print(accuracy_score(y_test, predictions_svm))
print(confusion_matrix(y_test, predictions_svm))
print(classification_report(y_test, predictions_svm))
```
