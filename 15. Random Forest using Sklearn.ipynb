```python
import numpy as np
import pandas as pd
```

```python
titanic = pd.read_csv("Titanic-Dataset.csv")
print(titanic.shape)
titanic.head()
```

```python
df = titanic.copy()
```

```python
# removing useless columns - name, ticket, passenger id
df.drop(["PassengerId", "Name", "Ticket", "Fare", "Cabin"], axis = 1, inplace = True)
df.head()
```

```python
# handling string based data
def get_num(s):
    if s == "male":
        return 1
    else:
        return 0
df["Gender"] = df["Sex"].apply(get_num)
del df["Sex"]
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
mean_age_sur = df[df.Survived == 1].Age.mean()
mean_age_not_sur = df[df.Survived == 0].Age.mean()

df.Age.fillna(mean_age_sur, inplace = True)
df.Age.fillna(mean_age_not_sur, inplace = True)
df.head()
```

```python
def getNumberEmbarked(str):
    if str=='C':
        return 1
    elif str=='Q':
        return 2
    else:
        return 3
df['Embarked']= df["Embarked"].apply(getNumberEmbarked)
df.head()
```

```python
from sklearn import model_selection
X = df.iloc[:,0:df.shape[1]-1]
Y = df.iloc[:,-1]

X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y, random_state = 1)
```

```python
Y_train.isnull().sum()
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
clf.fit(X_train, Y_train)
Y_pred_l = clf.predict(X_test)
clf.score(X_train,Y_train), clf.score(X_test,Y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth = 6, random_state=0)
clf.fit(X_train, Y_train)
clf.score(X_train,Y_train),clf.score(X_test,Y_test)
```
